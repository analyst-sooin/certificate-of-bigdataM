```python
# 시험환경 세팅 (코드 변경 X)
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

def exam_data_load(df, target, id_name="", null_name=""):
    if id_name == "":
        df = df.reset_index().rename(columns={"index": "id"})
        id_name = 'id'
    else:
        id_name = id_name
    
    if null_name != "":
        df[df == null_name] = np.nan
    
    X_train, X_test = train_test_split(df, test_size=0.2, random_state=2021)
    
    y_train = X_train[[id_name, target]]
    X_train = X_train.drop(columns=[target])

    
    y_test = X_test[[id_name, target]]
    X_test = X_test.drop(columns=[target])
    return X_train, X_test, y_train, y_test 
    
df = pd.read_csv("../input/pima-indians-diabetes-database/diabetes.csv")
X_train, X_test, y_train, y_test = exam_data_load(df, target='Outcome')

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## 라이브러리 및 데이터 불러오기

```python
# 라이브러리
import pandas as pd

x = X_train.copy()
y = y_train.copy()
test = X_test.copy()
```

```python
# EDA
# info, describe, null
x.describe()
```

```python
scale_var =['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
```

```python
#피처엔지니어링ㄱㄱ

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x[scale_var] = scaler.fit_transform(x[scale_var])
test[scale_var] = scaler.fit_transform(test[scale_var])
```

```python
id = test.pop('id')
```

```python
# 머신러닝 ㄱㄱ

from sklearn.ensemble import RandomForestClassifier

model =RandomForestClassifier()
model.fit(x,y['Outcome'])
pred = model.predict(test)
```
