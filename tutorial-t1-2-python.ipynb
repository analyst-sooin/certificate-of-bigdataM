## 문제3
- 데이터셋(basic1.csv)의 'age'컬럼의 이상치를 더하시오!
- 단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함

```python
import numpy as np
import pandas as pd
df = pd.read_csv('../input/bigdatacertificationkr/basic1.csv')
```

```python
# 이상치 정의

age_mean = df['age'].mean() #age 평균
age_sv = np.std(df['age'])
outliar_low = age_mean - (age_sv * 1.5)
outliar_high = age_mean + (age_sv * 1.5)
print(outliar_low, outliar_high)
```

```python
high_data = df[df['age'] >= outliar_high]
low_data = df[df['age'] <= outliar_low]
```

```python
high_sum = high_data['age'].sum()
low_sum = low_data['age'].sum()
print(high_sum + low_sum)
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

df = pd.read_csv('../input/bigdatacertificationkr/basic1.csv')
```

```python
std = np.std(df['age']) * 1.5
mean = df['age'].mean()

min_out = mean - std
max_out = mean + std
print(min_out, max_out)
```

```python
# 이상치 age합
df[(df['age']>max_out)|(df['age']<min_out)]['age'].sum()

# 다르게 작성방법
# df.loc[(df['age'] > max)]['age'].sum() + df.loc[(df['age']< min)]['age'].sum()
```
